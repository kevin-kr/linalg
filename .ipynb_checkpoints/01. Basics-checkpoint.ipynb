```python
import numpy as np
import matplotlib.pyplot as plt
```

## Vectors

```python
ax = plt.axes()
ax.arrow(0, 0, .4, .4, head_width=.05, head_length=.1, fc='k', ec='k')
plt.show()
```

### 2D vectors

```python
u = np.array([3, 4])
v = np.array([5, 2])
```

```python
x_coords, y_coords = zip(u, v)
plt.scatter(x_coords, y_coords, color=['r', 'b'])
plt.axis([0, 8, 0, 6])
plt.grid()
plt.show()
```

```python
def plot_vector2d(vector2d, origin=[0, 0], **options):
    return plt.arrow(origin[0], origin[1],
                     vector2d[0], vector2d[1],
                     head_width=.2, head_length=.3, length_includes_head=True,
                     **options)
```

```python
plot_vector2d(u, color='r')
plot_vector2d(v, color='b')
plt.axis([0, 8, 0, 6])
plt.grid()
plt.show()
```

### 3D vectors

```python
x = np.random.normal(-5, 5, 30)
y = np.random.normal(-5, 5, 30)
z = np.random.normal(-5, 5, 30)
```

```python
from mpl_toolkits.mplot3d import Axes3D

ax = plt.subplot(111, projection='3d')
ax.scatter(x, y, z)
plt.show()
```

```python
def plot_3dvectors(ax, vectors, z0, **options):
    
    for v in vectors:
        x, y, z = v
        ax.plot([x,x], [y,y], [z0, z], color='gray', linestyle='dotted', marker='.')
    
    x_coords, y_coords, z_coords = zip(*vectors)
    ax.scatter(x_coords, y_coords, z_coords, **options)
```

```python
ax = plt.subplot(111, projection='3d')

for i in range(3):
    ax.plot([x[i], x[i]],
            [y[i], y[i]],
            [0, z[i]],
            color='gray', linestyle='dotted', marker='.')
    
ax.scatter(x[0:3], y[0:3], z[0:3], color=('r', 'g', 'b'))
plt.show()
```

### Norm

```python
def vector_norm(vector):
    return sum([element**2 for element in vector])**.5

u = np.array([2, 3, 5])
vector_norm(u)
```

```python
import numpy.linalg as LA
LA.norm(u)
```

```python
u = np.array([4, 3])
v = np.array([3, 4])

r = LA.norm(u)
plt.gca().add_artist(plt.Circle((0,0), r, color='#DDDDDD'))
plot_vector2d(u, color='red')
plot_vector2d(v, color='red')
plt.axis([0, 6, 0, 6])
plt.grid()
plt.show()
```

### Addition

```python
u = np.array([2, 4])
v = np.array([3, 1])

plot_vector2d(u, color='r')
plot_vector2d(v, color='b')
plot_vector2d(v, origin=u, color='b', linestyle='dotted')
plot_vector2d(u, origin=v, color='r', linestyle='dotted')
plot_vector2d(u+v, color='g')
plt.axis([0, 6, 0, 6])
plt.text(1, 3, 'u', color='r', fontsize=18)
plt.text(4.2, 2.5, 'u', color='r', fontsize=18)
plt.text(2, .3, 'v', color='b', fontsize=18)
plt.text(3, 4.7, 'v', color='b', fontsize=18)
plt.text(2.2, 3, 'u+v', color='g', fontsize=18)
plt.grid()
plt.show()
```

### Zero, unit and normal vectors

```python
plt.gca().add_artist(plt.Circle((0,0),1,color='c'))
plt.plot(0, 0, "ko")
plot_vector2d(v / LA.norm(v), color="k")
plot_vector2d(v, color="b", linestyle=":")
plt.text(0.3, 0.3, "$\hat{u}$", color="k", fontsize=18)
plt.text(1.5, 0.7, "$u$", color="b", fontsize=18)
plt.axis([-1.5, 4, -1.5, 2])
plt.grid()
plt.show()
```

### Dot product

### Projection

```python
u = np.array([2,5])
v = np.array([3,1])

u_normalized = u / LA.norm(u)
proj = v.dot(u_normalized) * u_normalized

plot_vector2d(u, color='r')
plot_vector2d(v, color='b')

plot_vector2d(proj, color="k", linestyle=":")
plt.plot(proj[0], proj[1], "ko")

plt.plot([proj[0], v[0]], [proj[1], v[1]], "b:")

plt.text(1, 2, "$proj_u v$", color="k", fontsize=18)
plt.text(1.8, 0.2, "$v$", color="b", fontsize=18)
plt.text(0.8, 3, "$u$", color="r", fontsize=18)

plt.axis([0, 8, 0, 5.5])
plt.grid()
plt.show()

```
